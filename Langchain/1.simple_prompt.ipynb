```python
import os
from langchain_openai import ChatOpenAI
from langchain_core.prompts import PromptTemplate
from langchain_core.output_parsers import StrOutputParser
```

```python
os.environ["OPENAI_API_KEY"] = "gggg"

from openai import OpenAI
client = OpenAI()
```

```python
prompt = PromptTemplate.from_template("{question}")

# Temperature is a parameter that controls the randomness/creativity of the model’s responses.
# It ranges from 0.0 to 1.0 (sometimes higher):
# 0 → Deterministic, predictable, focused answers. The model will almost always give the same output for the same input.
# ~0.3–0.5 → Balanced: some creativity, but still fairly consistent.
# 0.7–1.0 → Highly creative, diverse, and sometimes less factual or consistent.

model = ChatOpenAI(model="gpt-4o-mini", temperature=0)
parser = StrOutputParser()

# Chain: prompt → model → parser
cahin = prompt | model | parser

# Runer
result = cahin.invoke({"question": "What is the capital of Pakistan"})
print(result)
```
