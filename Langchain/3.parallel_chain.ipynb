```python
import os
from langchain_openai import ChatOpenAI
from langchain_core.prompts import PromptTemplate
from langchain_core.runnables import RunnableParallel
from langchain_core.output_parsers import StrOutputParser
```

```python
os.environ["OPENAI_API_KEY"] = "hhhhh"

from openai import OpenAI
client = OpenAI()
```

```python
math_prompt = PromptTemplate.from_template("What is {number} squared?")
fact_prompt = PromptTemplate.from_template("tell me a fun fact about {number} ")
```

```python
model = ChatOpenAI(model="gpt-4o-mini", temperature=0)
```

```python
parser = StrOutputParser()
```

```python
math_chain = math_prompt | model | parser
fact_chain = fact_prompt | model | parser
```

```python
parallel_chain = RunnableParallel( 
    math=math_chain,
    fact=fact_chain
)
```

```python
result = parallel_chain.invoke({"number":13})
print(result)
```
